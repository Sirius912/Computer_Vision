## Problem 1
#### imread and imshow

```python
import cv2 as cv
import sys
import matplotlib.pyplot as plt

bus = cv.imread('bus.jpg')

if bus is None:
    sys.exit('File not found')

plt.imshow(bus)
plt.axis('off')
plt.title('bus')
plt.show()
```

## Problem 2
#### imwrite (color image -> gray image)

```python
bus_gray = cv.cvtColor(bus, cv.COLOR_BGR2GRAY)

cv.imwrite('bus_gray.jpg', bus_gray)

plt.subplot(1,2,1)
plt.imshow(bus)
plt.title('bus')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(bus_gray, cmap='gray')
plt.title('bus_gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Problem 3-1
#### Binary Image
#### - Color image -> binary image

```python
(thresh, bus_gray_thresh_30) = cv.threshold(bus_gray, 30, 255, cv.THRESH_BINARY)
(thresh, bus_gray_thresh_60) = cv.threshold(bus_gray, 60, 255, cv.THRESH_BINARY)

plt.subplot(1,2,1)
plt.title('bus_gray_thresh_30')
plt.imshow(bus_gray_thresh_30, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title('bus_gray_thresh_60')
plt.imshow(bus_gray_thresh_60, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Problem 3-2
#### Otsu Algorithm
#### - Optimization to obtain the best Threshold value

```python
(thresh, bus_gray_otsu) = cv.threshold(bus_gray, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)

print('The Best Threshold Value obtained by Otsu = ', thresh)

plt.imshow(bus_gray_otsu, cmap='gray')
plt.title('bus_gray_otsu')
plt.axis('off')
plt.show()
```

## Problem 3-3
#### Histogram

```python
histogram = cv.calcHist([bus_gray], [0], None, [256], [0,256])

plt.plot(histogram, color='b')
```

## Problem 4-1
#### RGB Channel
#### - In OpenCV -> BGR order!

```python
# red
bus_red = bus.copy()
bus_red[:,:,0] = 0      # Delete Blue
bus_red[:,:,1] = 0      # Delete Green
cv.imwrite('bus_red.jpg', bus_red)

# green
bus_green = bus.copy()
bus_green[:,:,0] = 0    # Delete Blue
bus_green[:,:,2] = 0    # Delete Red
cv.imwrite('bus_green.jpg', bus_green)

# blue
bus_blue = bus.copy()
bus_blue[:,:,1] = 0     # Delete Green
bus_blue[:,:,2] = 0     # Delete Red
cv.imwrite('bus_blue.jpg', bus_blue)

# Visualize
bus_list = [bus, bus_red, bus_green, bus_blue]
bus_title = ['bus', 'bus_red', 'bus_green', 'bus_blue']

for i in range(4):
    bus_rgb = cv.cvtColor(bus_list[i], cv.COLOR_BGR2RGB)
    plt.subplot(2,2,i+1)
    plt.imshow(bus_rgb)
    plt.title(bus_title[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## Problem 4-2
#### RGB Histogram
#### - In OpenCV -> BGR order!

```python
histogram_red = cv.calcHist([bus], [0], None, [256], [0,256])
histogram_green = cv.calcHist([bus], [1], None, [256], [0,256])
histogram_blue = cv.calcHist([bus], [2], None, [256], [0,256])

histogram_list = [histogram_red, histogram_green, histogram_blue]
color_list = ['r', 'g', 'b']

for i in range(3):
    plt.subplot(2,2,i+1)
    plt.plot(histogram_list[i], color=color_list[i])

plt.tight_layout()
plt.show()
```
